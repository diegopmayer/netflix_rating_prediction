```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('bmh')
plt.rcParams['figure.figsize'] = [20,9]
plt.rcParams['font.size'] = 18
```

```python
df = pd.read_csv("../dataset/bronze/dataset_video_games.csv")
```

```python
df.head()
```

```python
df['Genre'].head(150).hist(bins=15)
```

```python
sns.histplot(data=df.head(150), bins=10, y='Genre')
plt.title("Histogram of count of game per genre")
```

```python
df1 = df[df['Publisher'] == 'Nintendo']
df1 = df1[df1['Year'] >= 2012]
df1.head()
```

```python
sns.scatterplot(data=df1, x='Year', y='Global_Sales', hue='Publisher', palette="deep")
```

```python
aux = df[['EU_Sales', 'Publisher']].groupby(by='Publisher').sum(
    ).sort_values(by='EU_Sales', ascending=False).head(5).reset_index()
aux
```

```python
sns.barplot(data=aux, x='Publisher', y='EU_Sales')
```
